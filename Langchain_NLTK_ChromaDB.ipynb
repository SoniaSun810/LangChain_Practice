```python
# Initialize stemmer, lemmatizer, and stopwords
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
import string
def preprocess(text):
    """
    This function takes in a string of text, then performs the following:
    1. Tokenizes the sentence
    2. Stems the tokens
    3. Lemmatizes the tokens
    4. Removes stopwords and punctuations
    5. Returns a list of cleaned tokens
    """
    # Tokenize
    tokens = word_tokenize(text)

    # Stemming (you can choose to use either stemming or lemmatizing)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    # Lemmatizing
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Removing Stopwords and punctuations
    filtered_tokens = [token for token in lemmatized_tokens if token.lower() not in stop_words]
    filtered_tokens2 = [token for token in filtered_tokens if token not in string.punctuation]

    return filtered_tokens2
```

```python
import csv
def store_students_info_keywords_csv(students):
    """
    Store the student's UUID and key words tokens in a CSV file
    param: students: list of students (dict of student_uuid, key_words_tokens)
    """
    # Create or overwrite the CSV with headers
    with open('student_data_keywords.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["UUID", "background_key_words"])

    # Process each student and append their data to the CSV
    for student in students:
        student_uuid = student["student_uuid"]
        student["key_words_tokens"].sort()
        student_key_words_tokens = ','.join(student["key_words_tokens"])
        
        # Append to CSV
        with open('student_data_keywords.csv', 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([student_uuid, student_key_words_tokens])
```

```python
# Helper Function: Importing langChain document loaders and loading the data
from langchain.document_loaders.csv_loader import CSVLoader
def load_data_csv_file(file_path):
    loader = CSVLoader(file_path)
    data = loader.load()
    return data

from langchain.text_splitter import CharacterTextSplitter
def split_documents(documents, chunk_size=1000, chunk_overlap =0):
    text_splitter = CharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
    texts = text_splitter.split_documents(documents)
    return texts
```

```python
def process_students_info_populate_csv(file_path = 'selected_columns_only.csv'):
    # Process the whole dataset and save the results in a csv file
    students_data = load_data_csv_file(file_path)
    students_text = split_documents(students_data)
    students = []

    for student_text in students_text:
        text = student_text.page_content
        cleaned_tokens = preprocess(text)
        student_dict = get_student_info_uuid_keywords(cleaned_tokens)
        students.append(student_dict)
    
    print(len(students))
    print(students[0])
    store_students_info_keywords_csv(students)   
```

```python
import chromadb
from langchain.vectorstores import Chroma
import os
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(openai_api_key=os.getenv("OPENAI_API_KEY"))
students_data = load_data_csv_file('student_data_keywords.csv')
students_text = split_documents(students_data)

persist_directory = "/Users/sunfengnan/Documents/TA_Matching/LangChain_Practice"
client = chromadb.PersistentClient(path=persist_directory)

vector_db = Chroma.from_documents(
    client=client,
    documents=students_text, 
    collection_name="ta_matching_students_info",
    embedding=embeddings,
    persist_directory=persist_directory,
)
print(f"Documents Loaded: {vector_db._collection.count()}")
```

```python
def get_similar_docs(db, query, k = 20, score = False):
    if score:
        similar_docs = db.similarity_search_with_score(query, k)
    else:
        similar_docs = db.similarity_search(query, k)
    return similar_docs
```
